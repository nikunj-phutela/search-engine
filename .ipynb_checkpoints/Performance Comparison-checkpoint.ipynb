```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from ast import literal_eval
import pickle
```

```python
with open('./Relevant Pickles/relevant_final-10.pickle',"rb") as f:
    relevant = pickle.load(f)
```

```python
with open('./Retrieved Pickles/RetrievedPhraseText-10.pickle',"rb") as f:
    retrievedphrase = pickle.load(f)
```

```python
with open('./Retrieved Pickles/RetrievedFreeText-10.pickle',"rb") as f:
    retrievedfree = pickle.load(f)
```

```python
queryfile="sample_queries"
inputfile= open(queryfile,"r")
q=[]
for line in (inputfile.read().split('\n')):
    q.append(line)
inputfile.close()
```

```python

def evaluationmeasure(retrieved):
    pnum = 0
    pden = 0
    rden=0

    for line in q:
    #Precision=Total number of documents retrieved that are relevant/Total number of documents that are retrieved.
    #Recall=Total number of documents retrieved that are relevant/Total number of relevant documents in the database.
        pnum+=len(set((retrieved[line][0])).intersection((relevant[line][0])))
        pden+=len(set(retrieved[line][0]))
        rden+=len(set(relevant[line][0]))
    p=pnum/pden
    r=pnum/rden
    f=2*p*r/(p+r)
    print("Precision: ",p)
    print("Recall: ",r)
    print("F measure: ",f)
```

```python
#Precision Recall F measure for free text
evaluationmeasure(retrievedfree)
```

```python
#Precision Recall F measure for Phrase text
evaluationmeasure(retrievedphrase)
```

```python
def plotPRcurve(retrieved,size):
    
    p=[]
    r=[]
    pnum = 0
    pden = 0
    rden=0

    for k in range(1,size):
        for line in q:
            #print(line)
            pnum+=len(set((retrieved[line][0][:k])).intersection((relevant[line][0])))
            pden+=len(set(retrieved[line][0][:k]))
            rden+=len(set(relevant[line][0]))
        p.append(pnum/pden)
        r.append(pnum/rden)
    #print(p)
    #print(r)
    plt.figure(figsize=(10, 5), dpi=100)
    plt.plot(r, p)
    plt.xlabel("Recall")
    plt.title('Precision Recall Curve')
    plt.ylabel("Precision")
    plt.show()
```

```python
plotPRcurve(retrievedfree,10)
```

```python
plotPRcurve(retrievedphrase,10)
```

```python
def timecomparision(retrieved):
    
    plt.figure(figsize=(10, 6), dpi=100)
    ind = np.arange(47) 
    width = 0.35       
    plt.bar(ind, list(retrieved.loc["time"][3:50]), width, label='Our Search Engine')
    plt.bar(ind + width, list(relevant.loc["time"])[3:50], width, label='Elastic search')
    plt.xlabel("Query ")
    plt.ylabel('Time')
    plt.title('Time taken')

    plt.legend(loc='upper right')
    plt.show()
```

```python
timecomparision(retrievedphrase)
```

```python
timecomparision(retrievedfree)
```
